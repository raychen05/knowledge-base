### SQL - Stream Downstream Table

```scala

%sql

CREATE MATERIALIZED VIEW ag_ra_search_analytics_data_dev.sandbox_v1_0.delta_sp_org_link AS
SELECT 
  sp.sp_id,
  dom.org_pguid
FROM ag_ra_search_analytics_data_dev.sandbox_v1_0.d_spmaster AS sp
JOIN ag_ra_search_analytics_data_dev.sandbox_v1_0.spmaster_publication_link AS apl
  ON sp.sp_id = apl.sp_id
JOIN ag_ra_search_analytics_data_dev.sandbox_v1_0.orgmaster_publication_link AS ompl
  ON apl.uid = ompl.uid AND apl.author_position = ompl.address_position
JOIN ag_ra_search_analytics_data_dev.sandbox_v1_0.d_orgmaster AS dom
  ON ompl.org_pguid = dom.org_pguid;


CREATE MATERIALIZED VIEW ag_ra_search_analytics_data_dev.sandbox_v1_0.delta_sp_org_link AS
SELECT DISTINCT
  sp.sp_id,
  a.org_pguid
FROM ag_ra_search_analytics_data_dev.sandbox_v1_0.d_spmaster AS sp
JOIN (
    SELECT 
        ot.org_pguid,
        om.institution_key
    FROM ag_content_ims_dev.acs_entity.d_orgmaster ot
    JOIN ag_content_ims_dev.acs_wos.d_daisng_ranked_affiliation om
    ON om.institution_key  = ot.org_pguid) AS a
ON a.org_pguid = ot.org_pguid



CREATE MATERIALIZED VIEW ag_ra_search_analytics_data_dev.sandbox_v1_0.delta_sp_ut_link AS
SELECT 
  dom.org_pguid,
  ompl.uid
FROM ag_ra_search_analytics_data_dev.sandbox_v1_0.d_orgmaster AS dom
JOIN ag_ra_search_analytics_data_dev.sandbox_v1_0.orgmaster_publication_link AS ompl
  ON dom.org_pguid = ompl.org_pguid





CREATE MATERIALIZED VIEW ag_ra_search_analytics_data_dev.sandbox_v1_0.delta_sp_ut_link AS
SELECT 
  sp.sp_id,
  apl.uid
FROM ag_ra_search_analytics_data_dev.sandbox_v1_0.d_spmaster AS sp
JOIN ag_ra_search_analytics_data_dev.sandbox_v1_0.spmaster_publication_link AS apl
  ON sp.sp_id = apl.sp_id


```
