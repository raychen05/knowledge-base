```python
sealed trait ACS {
  def tableName: String
  def tableType: String
  def fullName: String = AcsResolver.getName(this)
}

object ACS {
 
  // gold_entity tables
  case object DAlmaOpenAccess extends ACS { val tableName = "d_alma_openaccess"; val tableType = "entity" }
  case object DAlmaSubscriptions extends ACS { val tableName = "d_alma_subscriptions"; val tableType = "entity" }
  case object DEsiAuthor extends ACS { val tableName = "d_esi_author"; val tableType = "entity" }
  case object DEsiCountry extends ACS { val tableName = "d_esi_country"; val tableType = "entity" }
  case object DEsiInstitution extends ACS { val tableName = "d_esi_institution"; val tableType = "entity" }
  case object DEsiJournal extends ACS { val tableName = "d_esi_journal"; val tableType = "entity" }
  case object DFundingOrganization extends ACS { val tableName = "d_funding_organization"; val tableType = "entity" }
  case object DGrantmasterDataitem extends ACS { val tableName = "d_grantmaster_dataitem"; val tableType = "entity" }
  case object DGrantmasterGrantid extends ACS { val tableName = "d_grantmaster_grantid"; val tableType = "entity" }
  case object DJcrJournals extends ACS { val tableName = "d_jcr_journals"; val tableType = "entity" }
  case object DNuts extends ACS { val tableName = "d_nuts"; val tableType = "entity" }
  case object DNutsCode extends ACS { val tableName = "d_nuts_code"; val tableType = "entity" }
  case object DOrcidRid extends ACS { val tableName = "d_orcid_rid"; val tableType = "entity" }
  case object DOrgmaster extends ACS { val tableName = "d_orgmaster"; val tableType = "entity" }
  case object DOrgmasterAssociation extends ACS { val tableName = "d_orgmaster_association"; val tableType = "entity" }
  case object DParentChild extends ACS { val tableName = "d_parent_child"; val tableType = "entity" }
  case object DSpmaster extends ACS { val tableName = "d_spmaster"; val tableType = "entity" }
  case object FIpInstitution extends ACS { val tableName = "f_ip_institution"; val tableType = "entity" }
  case object FIpInstitutionReputation extends ACS { val tableName = "f_ip_institution_reputation"; val tableType = "entity" }
  case object FJcrJournal extends ACS { val tableName = "f_jcr_journal"; val tableType = "entity" }
  case object FJcrJournalCategory extends ACS { val tableName = "f_jcr_journal_category"; val tableType = "entity" }
  case object FJcrJournalJci extends ACS { val tableName = "f_jcr_journal_jci"; val tableType = "entity" }
  case object FundingOrgPublicationLink extends ACS { val tableName = "funding_org_publication_link"; val tableType = "entity" }
  case object GrantmasterOrgmasterLink extends ACS { val tableName = "grantmaster_orgmaster_link"; val tableType = "entity" }
  case object GrantmasterPublicationLink extends ACS { val tableName = "grantmaster_publication_link"; val tableType = "entity" }
  case object JournalPublisherLink extends ACS { val tableName = "journal_publisher_link"; val tableType = "entity" }
  case object OrcidRidPublicationLink extends ACS { val tableName = "orcid_rid_publication_link"; val tableType = "entity" }
  case object OrgmasterHeirarchyPublicationLink extends ACS { val tableName = "orgmaster_heirarchy_publication_link"; val tableType = "entity" }
  case object OrgmasterHierarchyPublicationLink extends ACS { val tableName = "orgmaster_hierarchy_publication_link"; val tableType = "entity" }
  case object OrgmasterPublicationLink extends ACS { val tableName = "orgmaster_publication_link"; val tableType = "entity" }
  case object SpmasterPublicationLink extends ACS { val tableName = "spmaster_publication_link"; val tableType = "entity" }


  // gold_wos / gold_pprn tables
  case object AddressPublicationLink extends ACS { val tableName = "address_publication_link"; val tableType = "collection" }
  case object AuthorPublicationLink extends ACS { val tableName = "author_publication_link"; val tableType = "collection" }
  case object CategoryPublicationLink extends ACS { val tableName = "category_publication_link"; val tableType = "collection" }
  case object CountryPublicationLink extends ACS { val tableName = "country_publication_link"; val tableType = "collection" }
  case object CountryTerritoryLink extends ACS { val tableName = "country_territory_link"; val tableType = "collection" }
  case object DAlcCompleteLabels extends ACS { val tableName = "d_alc_complete_labels"; val tableType = "collection" }
  case object DArticleFlag extends ACS { val tableName = "d_article_flag"; val tableType = "dataset" }
  case object DArticleFlagWoscore extends ACS { val tableName = "d_article_flag_woscore"; val tableType = "collection" }
  case object DArticleIdentifiers extends ACS { val tableName = "d_article_identifiers"; val tableType = "collection" }
  case object DArticleMetrics extends ACS { val tableName = "d_article_metrics"; val tableType = "dataset" }
  case object DArticleMetricsWoscore extends ACS { val tableName = "d_article_metrics_woscore"; val tableType = "collection" }
  case object DArticleTotalCites extends ACS { val tableName = "d_article_total_cites"; val tableType = "dataset" }
  case object DArticleTotalCitesWoscore extends ACS { val tableName = "d_article_total_cites_woscore"; val tableType = "collection" }
  case object DArticleType extends ACS { val tableName = "d_article_type"; val tableType = "collection" }
  case object DArticleTypePrecedence extends ACS { val tableName = "d_article_type_precedence"; val tableType = "collection" }
  case object DAuthor extends ACS { val tableName = "d_author"; val tableType = "collection" }
  case object DCategory extends ACS { val tableName = "d_category"; val tableType = "collection" }
  case object DCitation extends ACS { val tableName = "d_citation"; val tableType = "collection" }
  case object DCitationFlag extends ACS { val tableName = "d_citation_flag"; val tableType = "collection" }
  case object DCitationPatent extends ACS { val tableName = "d_citation_patent"; val tableType = "dataset" }
  case object DCitationPatentWoscore extends ACS { val tableName = "d_citation_patent_woscore"; val tableType = "collection" }
  case object DCitationPatentFlag extends ACS { val tableName = "d_citation_patent_flag"; val tableType = "dataset"}
  case object DCitationPatentsFlag extends ACS { val tableName = "d_citation_patents_flag"; val tableType = "collection" }
  case object DCountry extends ACS { val tableName = "d_country"; val tableType = "collection" }
  case object DDaisngRankedAffiliation extends ACS { val tableName = "d_daisng_ranked_affiliation"; val tableType = "dataset" }
  case object DDaisngRankedAffiliationWoscore extends ACS { val tableName = "d_daisng_ranked_affiliation_woscore"; val tableType = "collection" }
  case object DDomesticCollab extends ACS { val tableName = "d_domestic_collab"; val tableType = "collection" }
  case object DDomesticInternationalCollab extends ACS { val tableName = "d_domestic_international_collab"; val tableType = "dataset" }
  case object DDomesticInternationalCollabWoscore extends ACS { val tableName = "d_domestic_international_collab_woscore"; val tableType = "collection" }
  case object DDomesticInternationalCollabtype extends ACS { val tableName = "d_domestic_international_collabtype"; val tableType = "collection" }
  case object DEdition extends ACS { val tableName = "d_edition"; val tableType = "collection" }
  case object DEmtCompleteLabels extends ACS { val tableName = "d_emt_complete_labels"; val tableType = "collection" }
  case object DEsciCitations extends ACS { val tableName = "d_esci_citations"; val tableType = "collection" }
  case object DEsciYearwiseCitations extends ACS { val tableName = "d_esci_yearwise_citations"; val tableType = "collection" }
  case object DGrantDataitem extends ACS { val tableName = "d_grant_dataitem"; val tableType = "collection" }
  case object DJournal extends ACS { val tableName = "d_journal"; val tableType = "collection" }
  case object DJournalAcs extends ACS { val tableName = "d_journal_acs"; val tableType = "collection" }
  case object DJournalPublisher extends ACS { val tableName = "d_journal_publisher"; val tableType = "collection" }
  case object DLocation extends ACS { val tableName = "d_location"; val tableType = "collection" }
  case object DNonEsciCitations extends ACS { val tableName = "d_non_esci_citations"; val tableType = "collection" }
  case object DNonEsciYearwiseCitations extends ACS { val tableName = "d_non_esci_yearwise_citations"; val tableType = "collection" }
  case object DOaArticle extends ACS { val tableName = "d_oa_article"; val tableType = "collection" }
  case object DOrganization extends ACS { val tableName = "d_organization"; val tableType = "collection" }
  case object DOrganizationCitationCount extends ACS { val tableName = "d_organization_citation_count"; val tableType = "dataset" }
  case object DOrganizationCitationCountWoscore extends ACS { val tableName = "d_organization_citation_count_woscore"; val tableType = "collection" }
  case object DOrganizationCitationFlag extends ACS { val tableName = "d_organization_citation_flag"; val tableType = "collection" }
  case object DPublicationSourceCitationCount extends ACS { val tableName = "d_publication_source_citation_count"; val tableType = "dataset" }
  case object DPublicationSourceCitationCountWoscore extends ACS { val tableName = "d_publication_source_citation_count_woscore"; val tableType = "collection" }
  case object DPublicationSourceCitationFlag extends ACS { val tableName = "d_publication_source_citation_flag"; val tableType = "collection" }
  case object DPublisher extends ACS { val tableName = "d_publisher"; val tableType = "collection" }
  case object DResearcherCitationCount extends ACS { val tableName = "d_researcher_citation_count"; val tableType = "dataset" }
  case object DResearcherCitationCountWoscore extends ACS { val tableName = "d_researcher_citation_count_woscore"; val tableType = "collection" }
  case object DResearcherCitationsFlag extends ACS { val tableName = "d_researcher_citations_flag"; val tableType = "collection" }
  case object DState extends ACS { val tableName = "d_state"; val tableType = "collection" }
  case object DTerritory extends ACS { val tableName = "d_territory"; val tableType = "collection" }
  case object DYearwiseCitation extends ACS { val tableName = "d_yearwise_citation"; val tableType = "dataset" }
  case object DYearwiseCitationWoscore extends ACS { val tableName = "d_yearwise_citation_woscore"; val tableType = "collection" }
  case object DYearwiseCitationCount extends ACS { val tableName = "d_yearwise_citation_count"; val tableType = "dataset" }
  case object DYearwiseCitationCountWoscore extends ACS { val tableName = "d_yearwise_citation_count_woscore"; val tableType = "collection" }
  case object FArticleCountry extends ACS { val tableName = "f_article_country"; val tableType = "dataset" }
  case object FArticleCountryWoscore extends ACS { val tableName = "f_article_country_woscore"; val tableType = "collection" }
  case object FCategory extends ACS { val tableName = "f_category"; val tableType = "dataset" }
  case object FCategoryWoscore extends ACS { val tableName = "f_category_woscore"; val tableType = "collection" }
  case object FCountry extends ACS { val tableName = "f_country"; val tableType = "dataset" }
  case object FCountryWoscore extends ACS { val tableName = "f_country_woscore"; val tableType = "collection" }
  case object FEsciCountry extends ACS { val tableName = "f_esci_country"; val tableType = "collection" }
  case object FNonEsciCountry extends ACS { val tableName = "f_non_esci_country"; val tableType = "collection" }
  case object FPublication extends ACS { val tableName = "f_publication"; val tableType = "collection" }
  case object GrantFundingOrganizationLink extends ACS { val tableName = "grant_funding_organization_link"; val tableType = "collection" }
  case object JournalAcsPublicationLink extends ACS { val tableName = "journal_acs_publication_link"; val tableType = "collection" }
  case object JournalPublicationLink extends ACS { val tableName = "journal_publication_link"; val tableType = "collection" }
  case object NutsPublicationLink extends ACS { val tableName = "nuts_publication_link"; val tableType = "collection" }
  case object PublisherPublicationLink extends ACS { val tableName = "publisher_publication_link"; val tableType = "collection" }
  case object ReprintAddressPublicationLink extends ACS { val tableName = "reprint_address_publication_link"; val tableType = "collection" }
  case object StatePublicationLink extends ACS { val tableName = "state_publication_link"; val tableType = "collection" }
  case object TerritoryPublicationLink extends ACS { val tableName = "territory_publication_link"; val tableType = "collection" }

}

object AcsResolver {
  private def getWidget(name: String, default: String): String = {
    try {
      val value = dbutils.widgets.get(name)
      if (value == null || value.isEmpty) default else value
    } catch {
      case _: Throwable => default
    }
  }

  private val environment = getWidget("environment", "dev")
  private val version = getWidget("version", "v1_0")
  private val collection = getWidget("collection", "wos")
 // private val dataset = getWidget("dataset", "woscore")
  private val catalog = s"ag_ra_search_analytics_data_${environment}"
  private var dataset = "" 

  def dataset(dataSet: String = ""): Unit ={
      dataset = if (dataSet == "woscore") "_woscore" else ""
  }

  def getName(table: ACS): String = {
    table.tableType match {
      case "entity"     => s"${catalog}.gold_entity_${version}.${table.tableName}"
      case "collection" => s"$catalog.gold_${collection}_$version.${table.tableName}"
      case "dataset"    => s"$catalog.gold_${collection}_$version.${table.tableName}${dataset}"
    }
  }
}
```
